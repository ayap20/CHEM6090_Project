# Exploratory data analysis for Henry's law constants

```python
# importing modules that will be used
import math
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('hcp_features.csv')
```

```python
# taking log of hcp values to make scale better
df['log Hcp'] = [math.log(x) for x in df['Hcp']]
```

```python
# histogram of hcp values to see spread of data
sns.histplot(data=df, x='log Hcp').set_title('Hcp distribution')
sns.despine()
plt.savefig('hcp_distribution.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Hcp vs tpsa
sns.scatterplot(data=df, x='tpsa', y='log Hcp').set_title('The effect of tpsa on log Hcp')
```
